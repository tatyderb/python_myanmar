### Последовательности дат

`pd.date_range(start, end, periods, freq)`

* start - начало периода генерации дат
* end - конец периода генерации дат
* periods - количество дат для генерации
* freq - период, с которым даты будут генерироваться

```python
# 15 дат от 2020-09-06 до 2023-05-18
pd.date_range('2020-09-06', '2023-05-18', periods=15)
```

```python
# нужны только даты, время не нужно
pd.date_range('2020-09-06', '2023-05-18', periods=15).normalize()
```

```python
# через каждые 2 дня и 6 часов
pd.date_range('2020-09-06 15:00', '2023-05-18', freq='2d 6h')
```

```python
# возьмем из набора дат первые periods штук
pd.date_range('2020-09-06 15:00', freq='2d 6h', periods=15)
```

#### Сокращения для freq

| Сокращение | Расшифровка | Сокращение | Расшифровка |
|----|----|----|----|
| B | Последний рабочий день в неделе | QS | Начало квартала |
| C | Пользовательский последний день в неделе | BQS | Первый рабочий день квартала |
| D | День | A, Y | Конец года |
| W | Неделя | BA, BY | Последний рабочий день в году |
| M | Конец месяца | AS, YS | Начало года |
| SM | 15 число месяца | BAS, BYS | Первый рабочий день в году |
| BM | Последний рабочий день в месяце | BH | Рабочий час (от 9:00 до 17:00) |
| CBM | Пользовательский конец месяца | H | Час |
| MS | Первое число месяца | T, min | Минута |
| SMS | 1 и 15 число месяца | S | Секунда |
| BMS | Первый рабочий день месяца | L, ms | Миллисекунда |
| CBMS | Пользовательский первый день месяца | U, us | Микросекунда |
| Q | Конец квартала | N | Наносекунда |
| BQ | Последний рабочий день квартала |  |  |

```python
pd.date_range('2020-08-06 15:00', '2023-05-18', freq='M')
```

## Метод shift

Если не последний день месяца, а пятый? Сдвинем на 5 дней последовательность.

```python
pd.date_range('2020-08-06 15:00', '2023-05-18', freq='MS').shift(5, 'd')
```

```python
# можно применить сдвиг несколько раз
(pd.date_range('2020-08-06 15:00', '2023-05-18', freq='Y')
    .shift(5, 'M')
    .shift(-2, 'h')
)
```

# Визуализация таблицы

* pd.options.display.max_columns - сколько столбцов показывать, остальные ... (None - все)
* pd.options.display.max_rows - сколько строк показывать, остальные ... (None - все)

Создадим большую таблицу и будем менять режимы отображения.

Создаем таблицу в которой 40 колонок и 100 строк:

```python
n_cols = 40
n_rows = 100

df = pd.DataFrame(np.arange(n_cols * n_rows).reshape(-1, n_rows).T,
            columns=[f'Колонка {i + 1}' for i in range(n_cols)])
df
```


Максимальное количество отобржаемых колонок



```python
pd.options.display.max_columns
```

Показаны 10 первых и 10 последних колонок. Это значение можно изменить.

```python
pd.options.display.max_columns = 30
df
```

```python
# None - отображать все колонки
pd.options.display.max_columns = None
df
```

```python
pd.options.display.max_columns = 7
pd.options.display.max_rows = 8
df
```

## df.sample(n=1) - случайные строки

* n = 1 - количество строк
* frac = 0.1 - доля данных по указанной оси (не употреблять вместе с n)
* axis = None - ось, 0 or 'index', 1 or 'columns', None default
* random_state - seed для генерации псевдослучайной последовательности (чтобы повторить ту же выборку)
* replace = True - разрешить тем же строкам попадать в выборку несколько раз, False - заменять одинаковые строки другими; по умолчанию False.

```python
df.sample()
```

```python
# запустите несколько раз, посмотрите, что получаем разные наборы строк
df.sample(n=5)
```

```python
df.sample(frac=0.05)  # 5% данных показать
```

### Случайное неслучайно

seed для генерации последовательности

```python
df.sample(n=5, random_state=2)
```

```python
# запускаем сколько угодно раз, получаем тот же набор строк указанного размера
df.sample(n=5, random_state=2)
# то же самое, записанное через явный вызов random_seed
# np.random.seed(2)
# df.sample(3)
```

```python
# Строка 74 повторяется
df.sample(10, replace=True, random_state=42)
```
